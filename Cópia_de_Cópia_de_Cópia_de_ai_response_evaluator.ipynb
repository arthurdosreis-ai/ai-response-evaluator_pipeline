```python
dataset = [
    {
        "prompt": "Why does the body have different smells?",
        "response_A": "Because of bacteria and sweat glands.",
        "response_B": "Different parts of the body produce different substances and bacteria create odors."
    }
]
```

```python
results = []

def evaluate_response(response):
    score = {
        "relevance": 0,
        "clarity": 0,
        "completeness": 0
    }

    if "body" in response.lower() or "ai" in response.lower() or "sleep" in response.lower():
        score["relevance"] = 2
    else:
        score["relevance"] = 1

    if len(response.split()) > 6:
        score["clarity"] = 2
    else:
        score["clarity"] = 1

    if len(response) > 50:
        score["completeness"] = 2
    else:
        score["completeness"] = 1

    return score


for item in dataset:
    score_A = evaluate_response(item["response_A"])
    score_B = evaluate_response(item["response_B"])

    total_A = sum(score_A.values())
    total_B = sum(score_B.values())

    if total_A > total_B:
        winner_auto = "A"
    elif total_B > total_A:
        winner_auto = "B"
    else:
        winner_auto = "Tie"

    human_choice = "B"
    agreement = winner_auto == human_choice

    result = {
        "prompt": item["prompt"],
        "score_A": score_A,
        "score_B": score_B,
        "total_A": total_A,
        "total_B": total_B,
        "winner_auto": winner_auto,
        "human_choice": human_choice,
        "agreement": agreement
    }

    results.append(result)

print(results)
total = len(results)
agreements = sum(1 for r in results if r["agreement"])

agreement_rate = agreements / total

print("Agreement rate:", agreement_rate)
```

```python
import json

with open("results.json", "w") as f:
    json.dump(results, f, indent=4)

print("Arquivo salvo!")
```
